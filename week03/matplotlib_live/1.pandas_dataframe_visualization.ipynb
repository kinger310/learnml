### 1.基本绘图 -- 折线图
Series和DataFrame上的这个功能只是使用matplotlib库的plot()方法的简单包装实现。

```python
df = pd.DataFrame(np.random.randn(10, 4))
df.plot()
```

我们可以使用`x`和`y`关键字绘制一列与另一列。

绘图方法允许除默认线图之外的少数绘图样式。 这些方法可以作为plot()的kind关键字参数提供。这些包括 -

* `bar`或`barh`为条形
* `hist`为直方图
* `boxplot`为盒型图
* `area`为“面积”
* `scatter`为散点图

### 2.条形图

要生成一个堆积条形图，通过指定：`pass stacked=True`

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=list("abcd"))
df.plot.bar(figsize=(8,6))
```

```python
df.plot.bar(stacked=True, figsize=(8,6))
```

```python
data = [
    ['Strategy 1', 100, 101, 140, 70, 54, 25, 100],
    ['Strategy 2', 105, 140, 144, 152, 35, 100, 44],
    ['Strategy 3', 120, 80, 160, 144, 20, 180, 60],
    ['Strategy 4', 90, 79, 162, 170, 99, 79, 16],
    ['Strategy 5', 118, 99, 137, 20, 181, 159, 13],
    ['Strategy 6', 18, 199, 117, 120, 131, 109, 43],
    ['Strategy 7', 92, 75, 127, 120, 21, 169, 33],
    ['Strategy 8', 118, 99, 137, 20, 181, 159, 13]
  ]
df = pd.DataFrame(data)
df = df.set_index(0)
df.index.name = "Sales Stratey"
df = df.T
df.plot.bar(stacked=True)
```

```python
import matplotlib
matplotlib.rc("figure", figsize=(8,6))
matplotlib.rc("font", size=10)
```

```python
df.plot.barh(stacked=True, figsize=(8,6), alpha=0.5)
```

要获得水平条形图，使用`barh()`方法

### 3.直方图
可以使用`plot.plot.hist()`方法绘制直方图。我们可以指定`bins`的数量值。

gaussian mixture model (GMM)

```python
df = pd.DataFrame({"a": np.random.randn(1000)+1, "b":np.random.randn(1000), "c": np.random.randn(1000)-1})
df.plot.hist(bins=30, alpha=0.5)
```

要为每列绘制不同的直方图，要用`df.hist()`

```python
df.hist()
```

### 5.区域块图形
可以使用`Series.plot.area()`或`DataFrame.plot.area()`方法创建区域图形

```python
df = pd.DataFrame(np.random.rand(10, 1), columns=list("A"))
df.plot.area()
```

### 6.散点图形
可以使用`DataFrame.plot.scatter()`方法创建散点图

看两个维度的关联程度

```python
df.corr()
```

```python
df = pd.DataFrame(np.random.randn(50, 4), columns=list("abcd"))
df.plot.scatter(x="a", y="b")
```

```python
s = np.random.randn(100, 1)
a = 100 + s
b = 10 + s
df = pd.DataFrame(np.concatenate([a, b], -1), columns=list("ab"))
df.plot.scatter(x="a", y="b")
```

```python
df.corr()
```

```python
import seaborn as sns
```

```python
sns.heatmap(df.corr())
```

### 7.饼状图
饼状图可以使用`DataFrame.plot.pie()`方法创建

```python
df = pd.DataFrame({'mass': [0.330, 4.87 , 5.97],
                   'radius': [2439.7, 6051.8, 6378.1]},
                  index=['Mercury', 'Venus', 'Earth'])
df.plot.pie(y="mass")
```
