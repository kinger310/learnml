## 基于pandas的内置可视化

### 1.基本绘图
Series和DataFrame上的这个功能只是使用matplotlib库的plot()方法的简单包装实现。

```python
%matplotlib inline
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(10,4),index=
                  pd.date_range('2018/12/18',  
   periods=10), columns=list('ABCD'))

df.plot().legend(bbox_to_anchor=(1.05,1))
```

```python
plt.rcParams("legend.bbox_to_anchor", (1.05,1))
```

如果索引由日期组成，则调用`gct().autofmt_xdate()`来格式化x轴，如上图所示。

我们可以使用`x`和`y`关键字绘制一列与另一列。

绘图方法允许除默认线图之外的少数绘图样式。 这些方法可以作为plot()的kind关键字参数提供。这些包括 -

* `bar`或`barh`为条形
* `hist`为直方图
* `boxplot`为盒型图
* `area`为“面积”
* `scatter`为散点图

### 2.条形图

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
print(df)
df.plot.bar(figsize=[8,6])
plt.legend(bbox_to_anchor=(0.1,0))
```

要生成一个堆积条形图，通过指定：`pass stacked=True`

```python
import pandas as pd
df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
print(df.head())
df.plot.bar(stacked=True)
```

要获得水平条形图，使用`barh()`方法

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])

df.plot.barh(stacked=True)
```

### 3.直方图
可以使用`plot.hist()`方法绘制直方图。我们可以指定`bins`的数量值。

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'a':np.random.randn(1000)+1,'b':np.random.randn(1000),'c':
np.random.randn(1000) - 1}, columns=['a', 'b', 'c'])
df.plot.hist(bins=20)
```

要为每列绘制不同的直方图，要用`df.plot()`

```python
import pandas as pd
import numpy as np

df=pd.DataFrame({'a':np.random.randn(1000)+1,'b':np.random.randn(1000),'c':
np.random.randn(1000) - 1}, columns=['a', 'b', 'c'])
df.hist(bins=20)
```

### 4.箱形图/盒型图/箱线图
`Boxplot`可以绘制调用`Series.box.plot()`和`DataFrame.box.plot()`或`DataFrame.boxplot()`来可视化每列中值的分布。

例如，这里是一个箱形图，表示对[0,1)上的统一随机变量的10次观察的五次试验。

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
df

# 绿线：中位数
# 上下蓝线：75% 25%分位数
# 上线边界(黑线)：边界 边界到箱子顶或者底的高度是1/2的箱子高度，如果没有则说明底已经是最小值
# describe()体现的分布去做了可视化
```

```python
df.iloc[2, 3] = 4
df.plot.box()
```

### 5.区域块图形
可以使用`Series.plot.area()`或`DataFrame.plot.area()`方法创建区域图形

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
df.plot.area()
```

### 6.散点图形
可以使用`DataFrame.plot.scatter()`方法创建散点图

看两个维度的关联程度

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.rand(50, 4), columns=['a', 'b', 'c', 'd'])
df.plot.scatter(x="a",y="b")
```

```python
df.corr() # pearson相关系数
```

```python
import seaborn as sns
sns.heatmap(df.corr())
```
